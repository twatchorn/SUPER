# Task
Create a user interface in Colab that allows users to upload `.pdb` files, input their name and email, and submit this information. Upon submission, display the uploaded filename and contact details to confirm successful data capture for a simulation workflow.

## Setup UI Environment

### Subtask:
Import necessary Python libraries like `ipywidgets` for creating interactive user interface elements within the Colab notebook.


**Reasoning**:
To import the `ipywidgets` library as `widgets` for creating interactive UI elements, I will use a code block.



````python
import ipywidgets as widgets
print("ipywidgets imported as widgets.")
````

## Create File Upload Widget

### Subtask:
Implement an ipywidgets.FileUpload widget to allow users to select and upload their .pdb files directly within the notebook interface. This step will also include a brief explanation of how to access the uploaded file's content.


**Reasoning**:
To implement the file upload widget, I will instantiate `ipywidgets.FileUpload` with the specified parameters to accept only single '.pdb' files and then display it.



````python
pdb_uploader = widgets.FileUpload(
    accept='.pdb',  # Accept only .pdb files
    multiple=False  # Allow only one file to be uploaded
)
display(pdb_uploader)
print("Upload the .pdb file for your system.")
````

## Create Contact Information Widgets

### Subtask:
Develop `ipywidgets.Text` input fields for users to enter their name and contact email address.


**Reasoning**:
To create interactive input fields for the user's name and email, I will instantiate two `ipywidgets.Text` widgets with appropriate descriptions and placeholders, and then display them.



````python
name_input = widgets.Text(
    description='Name:',
    placeholder='Enter your name'
)

email_input = widgets.Text(
    description='Email:',
    placeholder='Enter your email'
)

display(name_input, email_input)
print("Name and email input widgets created and displayed.")
````

## Create Submission Button and Handler

### Subtask:
Add an `ipywidgets.Button` that, when clicked, will trigger a Python function to collect the uploaded .pdb file data and the entered contact information. This function will be the entry point for your simulation logic.


**Reasoning**:
To implement the submission functionality, I will first create an `ipywidgets.Output` widget to capture and display results, then define a button and its click handler function as instructed, which will collect and display the uploaded file information and contact details.



````python
output_widget = widgets.Output()

submit_button = widgets.Button(
    description='Submit for Simulation',
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to submit your PDB file and contact information'
)

def on_submit_button_clicked(b):
    with output_widget:
        output_widget.clear_output()

        uploaded_filename = "No file uploaded"
        if pdb_uploader.value:
            uploaded_filename = next(iter(pdb_uploader.value))
            # file_content = pdb_uploader.value[uploaded_filename]['content']
            # For now, we just confirm the filename. Content processing would go here.

        name = name_input.value
        email = email_input.value

        print(f"--- Submission Details ---")
        print(f"Uploaded PDB File: {uploaded_filename}")
        print(f"Name: {name}")
        print(f"Email: {email}")
        print(f"--------------------------")

submit_button.on_click(on_submit_button_clicked)

display(submit_button, output_widget)
print("Submit button and output widget created and displayed.")
````
